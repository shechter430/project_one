```python
import pandas as pd
```

```python
life_expectancy = "resources/life_expectancy.csv"

data = pd.read_csv(life_expectancy)
```

```python
len(data)
```

```python
data.columns
```

```python
data.columns = ['Country', 'Year', 'Status', 'Life Expectancy', 'Adult Mortality (%)',
       'Infant Deaths (%)', 'Alcohol Consumpter per Capita (ltr)', 'Percentage Expenditure', 'Hepatitis B Vaccines (%)',
       'Measles Cases (%)', 'BMI', 'Under-5yo Deaths (%)', 'Polio Vaccines (%)', 'Total Expenditure (%)',
       'Diphtheria Vaccines (%)', 'HIV/AIDS Deaths Under 5yo (%)', 'GDP Per Capita ($)', 'Population',
       'Thinness  1-19 yrs', 'Thinness 5-9 yrs',
       'Income Composition of Resources', 'Schooling']
```

```python
data.columns
```

```python
data["Percentage Expenditure"] = data["Percentage Expenditure"].map("{:.2f}".format)
data["GDP Per Capita ($)"] = data["GDP Per Capita ($)"].map("{:.2f}".format)
data["Adult Mortality (%)"] = data["Adult Mortality (%)"] / 10
data["Adult Mortality (%)"] = data["Adult Mortality (%)"].map("{:.2f}".format)
data["Infant Deaths (%)"] = data["Infant Deaths (%)"] / 10
data["Infant Deaths (%)"] = data["Infant Deaths (%)"].map("{:.2f}".format)
data["Measles Cases (%)"] = data["Measles Cases (%)"] / 10
data["Measles Cases (%)"] = data["Measles Cases (%)"].map("{:.2f}".format)
data["Under-5yo Deaths (%)"] = data["Under-5yo Deaths (%)"] / 10
data["Under-5yo Deaths (%)"] = data["Under-5yo Deaths (%)"].map("{:.2f}".format)
data["Polio Vaccines (%)"] = data["Polio Vaccines (%)"].map("{:.0f}".format)
data["HIV/AIDS Deaths Under 5yo (%)"] = data["HIV/AIDS Deaths Under 5yo (%)"] / 10
data["HIV/AIDS Deaths Under 5yo (%)"] = data["HIV/AIDS Deaths Under 5yo (%)"].map("{:.2f}".format)
data["Hepatitis B Vaccines (%)"] = data["Hepatitis B Vaccines (%)"].map("{:.0f}".format)
data["Total Expenditure (%)"] = data["Total Expenditure (%)"].map("{:.2f}".format)
data["Diphtheria Vaccines (%)"] = data["Diphtheria Vaccines (%)"].map("{:.0f}".format)
data["Population"] = data["Population"].map("{:.0f}".format)
```
